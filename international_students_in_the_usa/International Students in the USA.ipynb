```python
import bar_chart_race as bcr
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./international_students_in_the_us.csv', index_col='Year')
```

```python
df.head()
```

```python
print (df.dtypes)
```

```python
# Replace NaN values
df.fillna(0, inplace=True)
```

```python
# Aggregate the date
df.iloc[:, 0:-1] = df.iloc[:, 0:-1].cumsum()
```
